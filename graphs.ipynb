```python
import matplotlib.pyplot as plt
import pylab
import numpy as np
x=np.linspace(0,20,1000)
y=np.sin(x)
pylab.plot(x,y,'g')
pylab.xlabel("This is x axis")
pylab.ylabel("This is y axis")
pylab.title("This is a sin graph")
```

```python
x = np.linspace(0, 20, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

pylab.plot(x, y1, color='orange', label = 'sine')
pylab.plot(x, y2, color='purple', label = 'cosine')
pylab.legend(loc = 'upper right')
pylab.ylim(-1.5, 2.0)
pylab.title("This is a sin and cos graph")
```

```python
x=np.linspace(0,20,100)
cosec=1/np.sin(x)
pylab.plot(x, color='pink', label = 'cosec')
pylab.legend(loc = 'upper right')
pylab.ylim(-1.5, 8.0)
pylab.title("This is a cosecant graph")
```

```python
x = np.linspace(0, 20, 1000)
y1 = np.sin(x)
y2 = 1/y1

pylab.plot(x, y1, color='orange', label = 'sine')
pylab.plot(x, y2, color='purple', label = 'cosec')
pylab.legend(loc = 'upper right')
pylab.ylim(-1.5, 2.0)
pylab.title("This is a sin and cosecant graph")
```
