```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mult(a,b):
    return a*b
def divide(a,b):
    return a/b
while(1):
    print("-------------------")
    print("Menu driven program")
    print("-------------------")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Exit")
    choice=int(input("Enter your choice: "))
    if choice==1:
        a=float(input("Enter the a value: "))
        b=float(input("Enter the b value: "))
        print("The sum of two digits is: ",add(a,b))
    elif choice==2:
        a=float(input("Enter the a value: "))
        b=float(input("Enter the b value: "))
        print("The difference of two digits is: ",sub(a,b))
    elif choice==3:
        a=float(input("Enter the a value: "))
        b=float(input("Enter the b value: "))
        print("The product of two digits is: ",mult(a,b))
    elif choice==4:
        a=float(input("Enter the a value: "))
        b=float(input("Enter the b value: "))
        print("The quotient of two digits is: ",divide(a,b))
    elif choice==5:
        break
    else:
        print("Invalid option! Try other numbers.")
```

```python
print("Enter the 5 elements of the list: ")
list1 = []
for i in range(5):
    list1.append(int(input()))
largest = list1[0]
for i in range(5):
    if largest < list1[i]:
        largest = list1[i]
secondlargest = list1[0]
for i in range(5):
    if secondlargest < list1[i]:
        if list1[i] != largest:
            secondlargest = list1[i]
print("The second largest element is: ")
print(secondlargest)
```

```python
n=int(input("Enter the value: "))
sum=1
for i in range(1,n+1):
    fact=1
    for j in range(1,i+1):
        fact*=j
    term=1/fact
    sum+=term
print("Sum= ",sum)
```
