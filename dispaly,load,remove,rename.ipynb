```python
#Write a menu driven program to implement the following
"""
(a)  Load the file from external directory
(b) display the contents of the file using with keyword
(c) rename the file.
(d) delete the file.
"""
import sys
import os
while(1):
    print("MENU DRIVEN PROGRAM")
    print("1. LOAD THE FILE FROM EXTERNAL MEMORY")
    print("2. DISPLAY THE CONTENTS OF THE FILE USING WITH KEYWORD")
    print("3. RENAME THE FILE")
    print("4. DELETE THE FILE")
    ch=int(input("ENTER YOUR CHOICE: "))
    if ch==1:
        try:
            path=os.path.abspath("sample.txt")
            sys.path.append(path)
        except:
            print("File doesnot exist")
        else:
            print("File opened successfully")
    elif ch==2:
          with open("abc1.txt","r") as f:
                for each in f:
                    print(each)
    elif ch==3:
        try:
            os.rename("Untitled9.ipynb","sqrt.ipynb")
        except:
            print("File not found")
        else:
            print("File renamed")
    elif ch==4:
        try: 
            os.remove("Untitled74.ipynb")
        except:
            print("File doesnot exist")
        else:
            print("File removed successfully")
    else:
        print("Invalid choice")
```
